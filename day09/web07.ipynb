## BeautifulSoup 모듈
* http 요청을 통해서 원하는 홈페이지에 있는
* 페이지 정보를 가져올 수 있는 모듈과 함께 사용함
* BeautifulSoup 모듈은 html 텍스트를 태그 기반으로 파싱해서
* 원하는 패턴을 파악하면 특정 정보를 가져올 수 있음
* 1) url  에 해당 주소를 할당하고 
* 2) requests 모듈을 사용해서 데이터를 가져오고
* 3) 그 정보를 html 변수에 할당함
* 4) soup 변수에 bs4 모듈의 BeautifulSoup 에 
*  현재 가져온 html 소스를 parser 와 함께 넘겨줌
* 5) div 태그 안에  class="tit3" 로 된 부분을 가져오기



```python
import requests
from bs4 import BeautifulSoup

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
req = requests.get(url)
html = req.text

# print(html)

soup = BeautifulSoup(html, 'html.parser')
# print(soup)

#print(soup.find('div', class_="tit3"))
print(soup.find_all('div', class_="tit3"))
# print(soup.find('div', class_="tit3").text)
```
