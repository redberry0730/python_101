###  naver 메인 페이지 정보 가져오기

* requests 모듈의 get(url) 함수를 사용해서
* naver 메인 페이지 정보 가져오기
* 검색결과시 나타나는 url을 적은 다음
* 'param=' 에 query 값으로 검색어를 입력하면
* 그 검색어에 해당하는 DATA 를 가져올 수 있음


```python
import requests

url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8'

# JSON 형태  query=python
param = {'query' : 'python'}
result = requests.get(url, params=param)
print(result.text)
```
